```python
#ACF and PACF Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
PD = pd.read_csv("C:/Users/deeks/Downloads/AirPassengers.csv")
PD
```

```python
plt.title('Airline Passengers Dataset', size=20)
plt.plot(PD["Passengers"])
```

```python
# First-order difference
PD['Passengers_Diff'] = PD['Passengers'].diff(periods=1)
PD = PD.dropna()
# Plot
plt.title('Airline Passengers dataset with First-order difference', size=20)
plt.plot(PD['Passengers'], label='Passengers')
plt.plot(PD['Passengers_Diff'], label='First-order difference', color='orange')
plt.legend();
```

```python
# Calculate autocorrelation
acf_values = acf(PD['Passengers_Diff'])
acf_values
```

```python
# Plot autocorrelation
plot_acf(PD['Passengers_Diff'], lags=30)
```

```python
# Calculate partial autocorrelation
pacf_values = pacf(PD['Passengers_Diff'])
pacf_values
# Plot partial autocorrelation
plot_pacf(PD['Passengers_Diff'], lags=30)
```
