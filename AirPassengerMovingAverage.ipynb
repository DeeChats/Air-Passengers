```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
#Importing Dataset
data=pd.read_csv("C:/Users/deeks/Downloads/AirPassengers.csv")
data
```

```python
#3 yearly Moving Average
data['SMA_3']= data['Passengers'].rolling(window=10).mean()
print(data['SMA_3'].tail(30))
```

```python
#line plot
data['Passengers'].plot(color='yellow',linewidth=3,figsize=(12,6))
data['SMA_3'].plot(color='blue',linewidth=3,figsize=(12,6))
#title and labels
plt.title('The Air Passengers',fontsize=20)
plt.xlabel('Month',fontsize=16)
plt.ylabel('Passengers',fontsize=16)
```
